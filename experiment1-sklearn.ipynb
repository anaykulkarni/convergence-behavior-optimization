```python
import numpy as np
import pandas as pd
import torch
from torch import nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset, random_split
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
import warnings
import json
```

```python
warnings.filterwarnings("ignore")
```

```python
wine = load_wine(as_frame=True)
dataset = pd.concat([wine.data, wine.target], axis=1)
print(f'Original Wine Dataset: Samples = {len(dataset)}, Labels = {dataset["target"].unique()}, Features = {len(dataset.columns)-1}')
dataset = dataset[dataset['target'] < 2]
print(f'Updated Wine Dataset: Samples = {len(dataset)}, Labels = {dataset["target"].unique()}, Features = {len(dataset.columns)-1}')
dataset
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset.iloc[:, :13], dataset.iloc[:, 13], test_size=0.2, random_state=42, shuffle=True)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict probabilities for training set
y_pred = model.predict_proba(X_train)  # Get probability estimates

# Compute training loss
train_loss = log_loss(y_train, y_pred)
print(f"Training Loss: {train_loss:.4f}")
```

```python
with open("losses-sklearn.json", "w") as file:
    json.dump(train_loss, file, indent=4)
```
